```python
greeting = 'Hello, World!'
print(greeting)

sentence = "It's a beautiful day!"
print(greeting + ' ' + sentence)
print(3 * (greeting + ' '))
# print(greeting * sentence)
```

```python
# slicing
my_string = 'Hello, World!'
print(len(my_string))
print(my_string[0])
print(my_string[7:12])
# read only
# my_string[0] = 'B'
# TypeError: 'str' object does not support item assignment
```

```python
# operations
my_string = ' Hello, World! , World! '
print(my_string.lower())
print(my_string.upper())
print(my_string.strip())
print(my_string.replace('World', 'Python'))
print(my_string.replace('World', 'Python', 1))
print(my_string.split(','))

# reverse replace
rev_str = my_string[::-1]
old = "World"[::-1]
new = "Python"[::-1]
rev_replaced = rev_str.replace(old, new, 1)
final_str = rev_replaced[::-1]
print(final_str)
```
