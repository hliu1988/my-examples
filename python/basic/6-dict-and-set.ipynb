```python
# Creating a dictionary
my_dict = {'apple': 1, 'banana': 2, 'cherry': 3}
print(my_dict)

# Accessing elements
print(my_dict['apple'])

# Adding elements
my_dict['date'] = 4
print(my_dict)

# Deleting elements
del my_dict['banana']
print(my_dict)

for key,val in my_dict.items():
    print(f"{key}: {val}")
```

```python
# Creating a set
my_set = {1, 2, 3, 4, 5}
print(my_set)

# Adding elements
my_set.add(6)
print(my_set)

# Removing elements
my_set.remove(3)
print(my_set)

# Set operations
another_set = {4, 5, 6, 7, 8}
print(my_set.union(another_set))
print(my_set.intersection(another_set))
```

```python
# Set difference
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
# set1 - set2
print(set1.difference(set2))  # Elements in set1 but not in set2

print(set1.union(set2))
print(set1.intersection(set2))
# Symmetric difference
print(set1.symmetric_difference(
    set2))  # Elements in either set1 or set2, but not in both

s = {1, 2}
# update()接收列表
s.update([3, 4, 4])  # 自动去重
print(s)  # 输出：{1, 2, 3, 4}
# union()接收元组
new_s = s.union((5, 6, 4, 4))  # 自动去重
print(new_s)  # 输出：{1, 2, 3, 4, 5, 6}
```

```python
# Shopping List Management Program
shopping_list = {}

def add_item(item, quantity):
    if item in shopping_list:
        shopping_list[item] += quantity
        print(f'Updated "{item}" quantity to {shopping_list[item]}.')
    else:
        shopping_list[item] = quantity
        print(f'"{item}" has been added to your shopping list with quantity {quantity}.')

def remove_item(item):
    if item in shopping_list:
        del shopping_list[item]
        print(f'"{item}" has been removed from your shopping list.')
    else:
        print(f'"{item}" not found in your shopping list.')

def display_list():
    if shopping_list:
        print("Your Shopping List:")
        total_items = 0
        for item, quantity in shopping_list.items():
            print(f"- {item}: {quantity}")
            total_items += quantity
        print(f"Total items: {total_items}")
    else:
        print("Your shopping list is currently empty.")

# Example Usage
add_item("Milk", 2)
add_item("Bread", 1)
display_list()
add_item("Eggs", 12)
display_list()
remove_item("Bread")
display_list()
```
