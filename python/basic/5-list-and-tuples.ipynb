```python
a = [1, -2, 3]
a.sort()
print(a)
print(a[::-1])

# 基本升序排序
nums = [5, 2, 9, 3, 7]
nums.sort()
print("升序排序后（原列表）:", nums)  # 输出 [2, 3, 5, 7, 9]

# 降序排序
nums.sort(reverse=True)
print("降序排序后（原列表）:", nums)  # 输出 [9, 7, 5, 3, 2]

# 按字符串长度排序
words = ["apple", "cat", "banana", "dog"]
words.sort(key=len)
print("按长度排序:", words)  # 输出 ['cat', 'dog', 'apple', 'banana']
```

```python
fruits = ["apple", "banana", "cherry"]

# Accessing List Items
print(fruits[1])

# Changing List Items
fruits[1] = "blackcurrant"
print(fruits)

# Looping Through a List
for fruit in fruits:
    print(fruit)

for i in range(0, len(fruits)):
    print(fruits[i])

for i,val in enumerate(fruits):
    print(i, val)
```

```python
# Creating a Tuple
fruits_tuple = ("apple", "banana", "cherry")
print(fruits_tuple)

fruits_tuple = ("apple", "banana", "cherry")
# Accessing Tuple Items
print(fruits_tuple[1])  # Output: banana

print("\norder1:")
for i in range(0, len(fruits_tuple)):
    print(fruits_tuple[i])

print("\norder2:")
for i,val in enumerate(fruits_tuple):
    print(i, val)

print("\nreverse order:")
for val in fruits_tuple[::-1]:
    print(val)
```

```python
# Shopping List Management Program
shopping_list = []

def add_item(item):
    if item not in shopping_list:
        shopping_list.append(item)
        print(f'"{item}" has been added to your shopping list.')
    else:
        print(f'"{item}" is already in your shopping list.')

def remove_item(item):
    if item in shopping_list:
        shopping_list.remove(item)
        print(f'"{item}" has been removed from your shopping list.')
    else:
        print(f'"{item}" not found in your shopping list.')

def display_list():
    if shopping_list:
        print("Your Shopping List:")
        for idx, item in enumerate(shopping_list, start=1):
            print(f"{idx}. {item}")
    else:
        print("Your shopping list is currently empty.")

# Example Usage
add_item("Milk")
add_item("Bread")
display_list()
remove_item("Milk")
display_list()
shopping_list.insert(0, "Eggs")
display_list()
```
