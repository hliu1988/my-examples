```python
def greet(name):
    print(f"Hello, {name}!")

greet("Elizabeth")
```

```python
def add(a, b):
    return a + b

result = add(5, 7)
print(result)
```

```python
def power(base, exponent=2):
    return base ** exponent

print(power(3))
print(power(3, 3))
print(power(base=2, exponent=3))
print(power(exponent=3, base=2))
```

```python
x = 10


def modify_global_var():
    # If you need to modify a variable from an outer scope inside a function, Python provides the global keyword:
    global x
    x = 20  # 没global，是local变量
    print(f"Inside the function, x is {x}")

modify_global_var()
print(f"Outside the function, x is also {x}")
```

```python
x = 10

def change_x(new_x):
    x = new_x  # local variable
    print(f"Inside the function, x is {x}")

change_x(20)
print(x)
```

```python
multiply = lambda x, y: x * y
print(multiply(5, 6))
```
