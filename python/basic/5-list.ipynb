## list.sort() vs. sorted(list)

- 如果你不需要保留原列表，且操作的是列表：用list.sort()（更高效，节省内存）；
- 如果你需要保留原数据，或要排序的不是列表（如元组、集合）：用sorted()（更灵活，适用范围广）。


```python
a = [1, -2, 3]
a.sort()
print(a)
print(a[::-1])

# 基本升序排序
nums = [5, 2, 9, 3, 7]
nums.sort()
print("升序排序后（原列表）:", nums)  # 输出 [2, 3, 5, 7, 9]

# 降序排序
nums.sort(reverse=True)
print("降序排序后（原列表）:", nums)  # 输出 [9, 7, 5, 3, 2]

# 按字符串长度排序
words = ["apple", "cat", "banana", "dog"]
words.sort(key=len)
print("按长度排序:", words)  # 输出 ['cat', 'dog', 'apple', 'banana']

## list.sort()

# 初始化列表
nums = [3, 1, 2, 5, 4]

# 调用sort()，原地排序
result = nums.sort()

# 原列表被修改（升序）
print("修改后的nums:", nums)  # 输出：[1, 2, 3, 4, 5]
# sort()无返回值（返回None）
print("sort()的返回值:", result)  # 输出：None

# 支持reverse参数（降序）
nums.sort(reverse=True)
print("降序后的nums:", nums)  # 输出：[5, 4, 3, 2, 1]

## sorted(list)
# 1. 排序列表（原列表不变）
nums = [3, 1, 2, 5, 4]
new_nums = sorted(nums)
print("原nums:", nums)        # 输出：[3, 1, 2, 5, 4]（不变）
print("排序后的新列表:", new_nums)  # 输出：[1, 2, 3, 4, 5]

# 2. 排序元组（返回列表）
t = (3, 1, 2)
sorted_t = sorted(t)
print("排序元组的结果:", sorted_t)  # 输出：[1, 2, 3]（列表类型）

# 3. 排序集合（返回列表）
s = {5, 2, 8, 1}
sorted_s = sorted(s)
print("排序集合的结果:", sorted_s)  # 输出：[1, 2, 5, 8]

# 4. 排序字符串（返回字符列表）
str1 = "bacde"
sorted_str = sorted(str1)
print("排序字符串的结果:", sorted_str)  # 输出：['a', 'b', 'c', 'd', 'e']

## key and reverse
# 示例：按字符串长度排序
words = ["apple", "banana", "cherry", "date"]

# 用sort()原地排序
words.sort(key=len, reverse=True)  # key=len表示按长度排序，reverse=True降序
print("sort()按长度排序:", words)  # 输出：['banana', 'cherry', 'apple', 'date']

# 用sorted()返回新列表
words2 = ["apple", "banana", "cherry", "date"]
new_words = sorted(words2, key=len, reverse=True)
print("原words2:", words2)        # 输出：["apple", "banana", "cherry", "date"]（不变）
print("sorted()按长度排序:", new_words)  # 输出：['banana', 'cherry', 'apple', 'date']
```

```python
# Shopping List Management Program
shopping_list = []

def add_item(item):
    if item not in shopping_list:
        shopping_list.append(item)
        print(f'"{item}" has been added to your shopping list.')
    else:
        print(f'"{item}" is already in your shopping list.')

def remove_item(item):
    if item in shopping_list:
        shopping_list.remove(item)
        print(f'"{item}" has been removed from your shopping list.')
    else:
        print(f'"{item}" not found in your shopping list.')

def display_list():
    if shopping_list:
        print("Your Shopping List:")
        for idx, item in enumerate(shopping_list, start=1):
            print(f"{idx}. {item}")
    else:
        print("Your shopping list is currently empty.")

# Example Usage
add_item("Milk")
add_item("Bread")
display_list()
remove_item("Milk")
display_list()
shopping_list.insert(0, "Eggs")
display_list()
```
