```python
res = 5 + 7
print(res)
res = 5.1 + 7.3
print(res)

print(int(res))
print(float(int(res)))
```

```python
# float
print(8 / 3)
# int
print(8 // 3) # floor
print((8 + 2) // 3)  # ceiling
# Note that floor division // behaves differently with negative numbers because it rounds toward negative infinity.
print(-8 // 3) # floor
print((-8 + 2) // 3) # ceiling
```

```python
# Modulus
# 判断奇数，偶数 x % 2 == 0
remainder = 8 % 3
print(remainder)

# Exponent
result = 2 ** 3
print(result)

# Sqrt
result = 2 ** 0.5
print(result)

# round
print(round(result, 2))
# 四舍五入的边界情况
print(round(2.675, 2))  # 注意：输出2.67（因浮点数二进制存储精度问题，并非2.68）
print(round(5.895, 2))  # 输出5.89（部分数值因精度问题会符合预期）
```

```python
# Working with complex numbers
z = 1 + 2j
print(z.real)  # Real part
print(z.imag)  # Imaginary part

z2 = 2 + 3j
sum_complex = z + z2
print(sum_complex)
```

```python
import math

# Examples using the math module
print(math.sqrt(16))  # prints 4.0
print(math.ceil(3.2)) # prints 4
print(math.pow(2, 3))
```
