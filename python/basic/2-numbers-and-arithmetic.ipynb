```python
# 普通写法：计算两次len(nums)
nums = [1, 2, 3, 4, 5, 6]
if len(nums) > 5:
    print(f"列表长度为 {len(nums)}，大于5")

# 海象运算符：仅计算一次len(nums)
if (length := len(nums)) > 5:
    print(f"列表长度为 {length}，大于5")

# # 普通写法
# while True:
#     user_input = input("请输入内容（空值退出）：")
#     if not user_input:
#         break
#     print(f"你输入了：{user_input}")

# # 海象运算符写法
# while (user_input := input("请输入内容（空值退出）：")):
#     print(f"你输入了：{user_input}")

words = ["apple", "cat", "banana", "dog", "elephant"]

# 普通写法：两次调用len(word)
result = [(word, len(word)) for word in words if len(word) > 3]

# 海象运算符：仅调用一次len(word)
result = [(word, length) for word in words if (length := len(word)) > 3]

print(result)  # 输出[('apple', 5), ('banana', 6), ('elephant', 8)]
```

```python
# float
print(8 / 3)
# int
print(8 // 3) # floor
print((8 + 2) // 3)  # ceiling
# Note that floor division // behaves differently with negative numbers because it rounds toward negative infinity.
print(-8 // 3) # floor
print((-8 + 2) // 3) # ceiling
```

```python
# Modulus
# 判断奇数，偶数 x % 2 == 0
remainder = 8 % 3
print(remainder)

# Exponent
result = 2 ** 3
print(result)

# Sqrt
result = 2 ** 0.5
print(result)

# round
print(round(result, 2))
# 四舍五入的边界情况
print(round(2.675, 2))  # 注意：输出2.67（因浮点数二进制存储精度问题，并非2.68）
print(round(5.895, 2))  # 输出5.89（部分数值因精度问题会符合预期）
```

```python
# Working with complex numbers
z = 1 + 2j
print(z.real)  # Real part
print(z.imag)  # Imaginary part

z2 = 2 + 3j
sum_complex = z + z2
print(sum_complex)
```
