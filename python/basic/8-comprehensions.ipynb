```python
# 推导式 通过在语法中嵌入循环和条件逻辑，提供了一种简洁且可读的方式来创建列表
# Basic list comprehension
squares = [x**2 for x in range(10)]
print(squares)

# List comprehension with condition
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print(even_squares)
```

```python
# Dictionary comprehension
square_dict = {x: x**2 for x in range(10)}
print(square_dict)

# Dictionary comprehension with condition
even_square_dict = {x: x**2 for x in range(10) if x % 2 == 0}
print(even_square_dict)
```

```python
# Set comprehension
unique_squares = {x**2 for x in range(10)}
print(unique_squares)
```

```python
# Nested list comprehension
matrix = [[j for j in range(5)] for i in range(3)]
print(matrix)
```

```python
# Flattening a nested list
# These techniques can substantially reduce the amount of code needed to handle more intricate data manipulations.
nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened_list = [item for sublist in nested_list for item in sublist]
print(flattened_list)
```

```python
# Generate a list comprehension in Python that produces a list of the squares of numbers from 1 to 20, but only include the numbers whose squares are divisible by 3.
squares_divisible_by_3 = [x**2 for x in range(1, 21) if (x**2) % 3 == 0]
print(squares_divisible_by_3)
```
