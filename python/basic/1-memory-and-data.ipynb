```python
# Converting a decimal number to binary in Python
num = 13
bin_rep = bin(num)
print(num, bin_rep)
print(num, bin_rep[2:])
```

```python
# Displaying the binary representation of a positive and negative integer
positive_num = 42
negative_num = -42
print(bin(positive_num))  # Output: 0b101010
print(bin(negative_num))
# short 的补码
print(bin(negative_num & 0xff))  # Output: -0b101010
```

```python
sum = 0
x = 0.1

for i in range(0, 10):
    sum += x

if sum == 1:
    print("sum is 1")
else:
    print("sum is not 1, but ", sum)

# Displaying the binary representation of a floating-point number
import struct

floating_num = 13.5
binary_representation = bin(struct.unpack('!I', struct.pack('!f', floating_num))[0])[2:]
print("Binary representation of", floating_num, ":", binary_representation)
```

```python
# Finding the ASCII and Unicode code points of a character
# char = 'A'
# ascii_value = ord(char)
# unicode_value = ord(char)
# print("ASCII value of", char, ":", ascii_value)
# print("Unicode value of", char, ":", unicode_value)

# 展示ASCII字符的ASCII值（同时也是Unicode码点）
char = '中'
# 先判断是否为ASCII字符（编码点0-127）
if ord(char) <= 127:
    ascii_value = ord(char)
    unicode_value = ord(char)
    print(f"ASCII value of '{char}': {ascii_value}")
    print(f"Unicode code point of '{char}': {unicode_value}")
else:
    print(f"'{char}' is not an ASCII character")
    print(f"Unicode code point of '{char}': {ord(char)}")

# 对比ASCII和非ASCII字符的编码
ascii_char = 'A'
non_ascii_char = '中'

# ASCII字符
print(f"\nCharacter '{ascii_char}':")
print(f"  ASCII value (Unicode code point): {ord(ascii_char)}")

# 非ASCII字符
print(f"\nCharacter '{non_ascii_char}':")
print(f"  Not an ASCII character")
print(f"  Unicode code point: {ord(non_ascii_char)}")

# Python 中的ord()函数并非专门用于获取 ASCII 值，它的作用是返回字符的Unicode 码点（code point）。对于非 ASCII 字符（如中文'中'、日文'あ'），ord()会返回其 Unicode 码点，而这些字符根本没有 ASCII 值。
# Python 3 中所有字符串都是Unicode 字符序列，不存在 “纯 ASCII 字符串” 的概念，ASCII 只是 Unicode 的一个子集。
```

```python
# Storing a string in Python
my_string = "Hello, World!"
print(f"The string is: {my_string}")
print(f"The memory address of the string is: {id(my_string)}")

# Memory addresses in Python data structures
my_list = [1, 2, 3]
my_dict = {'a': 1, 'b': 2}
print(f"Memory address of my_list: {id(my_list)}")
print(f"Memory address of my_dict: {id(my_dict)}")
```
