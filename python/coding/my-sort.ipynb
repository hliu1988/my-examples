```python
def quick_sort(nums):
    if not nums:
        return []
    pivot = nums[0]
    n = len(nums)
    left = [a for a in nums[1:] if a < pivot]
    right = [a for a in nums[1:] if a >= pivot]
    return quick_sort(left) + [pivot] + quick_sort(right)

nums = [5, 2, 9, 3, 7, 6, 1, 8, 4]
sorted_nums = quick_sort(nums)
print("排序结果:", sorted_nums)  # 输出 [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
def merge_sort(nums, low, high):
    if low >= high:
        return

    mid = (low + high) // 2
    merge_sort(nums, low, mid)
    merge_sort(nums, mid + 1, high)
    merge_in_place(nums, low, mid, high)

def merge_in_place(nums, low, mid, high):
    left = nums[low:mid + 1].copy()
    il, ih, i = 0, mid + 1, low
    while il < len(left) and ih <= high:
        if left[il] <= nums[ih]:
            nums[i] = left[il]
            il += 1
        else:
            nums[i] = nums[ih]
            ih += 1
        i += 1

    if il < len(left):  # 只用处理左边剩下的情况
        while il < len(left):
            nums[i] = left[il]
            i += 1
            il += 1

nums = [5, 2, 9, 3, 7, 6, 1, 8, 4]
merge_sort(nums, 0, len(nums) - 1)
print("排序结果:", nums)  # 输出 [1, 2, 3, 4, 5, 6, 7, 8, 9]
```
