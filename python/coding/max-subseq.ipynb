```python
# Kadane 算法（卡登算法）
# 这是解决该问题的最优解法，时间复杂度为 O(n)，空间复杂度为 O(1)。
def max_subarray_sum(nums):
    global_max = current_max = nums[0]
    for num in nums[1:]:
        # 选择：要么延续前一个子数组，要么以当前元素为新起点
        current_max = max(num, current_max + num)
        global_max = max(global_max, current_max)
    return global_max

# 测试示例
nums = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
print("最大子数组和（连续）:", max_subarray_sum(nums))  # 输出6
```

```python
def max_subsequence_sum(nums):
    if not nums:
        return 0
    # 筛选所有正数
    positive_nums = [num for num in nums if num > 0]
    if positive_nums:
        return sum(positive_nums)
    else:
        # 全为负数时返回最大元素
        return max(nums)

# 测试示例
nums1 = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
nums2 = [-3, -1, -2]
print("最大子序列和（非连续）nums1:", max_subsequence_sum(nums1))  # 输出1+4+2+1+4=12
print("最大子序列和（非连续）nums2:", max_subsequence_sum(nums2))  # 输出-1
```

```python
def max_subarray_elements(nums):
    if not nums:  # 处理空数组的边界情况
        return []
    current_max = global_max = nums[0]
    start = end = temp_start = 0  # temp_start记录当前子数组的临时起始索引

    for i in range(1, len(nums)):
        num = nums[i]  # 取出当前元素
        # 判断：是否以当前元素为新的子数组起点
        if current_max + num < num:
            current_max = num
            temp_start = i  # 更新临时起始索引为当前位置
        else:
            current_max += num

        # 如果当前子数组和超过全局最大值，更新全局信息
        if current_max > global_max:
            global_max = current_max
            start = temp_start  # 固定起始索引
            end = i  # 固定结束索引

    # 返回最大子数组的元素（切片是左闭右开，所以end+1）
    return nums[start:end+1]

# 测试
nums = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
print("最大子数组的元素:", max_subarray_elements(nums))  # 输出[4, -1, 2, 1]
```
